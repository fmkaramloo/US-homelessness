```python
import pandas as pd
```

```python
file = pd.read_csv("../master/datasets/shelters_final3.csv")
file.head()
```

```python
state_abbv = pd.read_csv("../master/total_pop/states_code.csv")
state_abbv.head()
```

```python
combine_df = pd.merge(file,state_abbv,on="State",how="left")
combine_df = combine_df.drop(columns=['Abbrev','Unnamed: 0'])
combine_df.head()
```

```python
combine_df = combine_df.sort_values("Code")
combine_df.head()
```

```python
sort_df = combine_df.reset_index(drop = True)
sort_df.head()
```

```python
sort_df.count()
```

```python
sort_df.to_csv("sort.csv",index = None, header=True)
```
