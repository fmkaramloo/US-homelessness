```python
sheets = range(2010, 2020)
dfs = []
for x in sheets:
    data_df = pd.read_excel(open('../US-homelessness/2007-2019_PIT_Counts_by_State.xlsx', 'rb'), sheet_name = str(x))
    colms=["Year"]
    renam = {}
    renam["Year"]="Year"
    for x1 in data_df.keys():
        if " ES " not in x1 and " TH " not in x1 and " SH " not in x1 and "Household" not in x1 and "Chronically" not in x1 and "Veterans" not in x1 and "Youth" not in x1:
#             colms.append(x1)
            renam[x1] = x1.replace(", "+str(x), "")
            colms.append(renam[x1])            
    data_df["Year"]= x
    data_df = data_df.rename(columns=renam)
    dfs.append(data_df[colms])
    print(x)
# print(renam)
```
