```python
# Dependencies
from census import Census
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
```

```python
# Census API Key
census_api_key = "85ac64b6b5a9c0901b00329d1ef41f0c53ccfc98"
```

```python

def census(c,year):
    census_data = c.acs5.get(("NAME", "B01003_001E"), {'for': 'state:*'})
    census_pd = pd.DataFrame(census_data)
    census_pd = census_pd.rename(columns={"NAME": "Name",
                                          "B01003_001E":"Total Population"})
    census_pd.to_csv(f"{year}.csv",index = None, header=True)
```

```python
years = range(2010,2019)
for year in years:
    c = Census(census_api_key, year)
    census(c,year)
```

```python
pop_files = sorted(glob("*.csv"))
pop_files
```

```python
pop_df = pd.concat((pd.read_csv(file).assign(Year=file) 
                       for file in pop_files),ignore_index=True)

pop_df = pop_df.replace("2010.csv","2010")
pop_df = pop_df.replace("2011.csv","2011")
pop_df = pop_df.replace("2012.csv","2012")
pop_df = pop_df.replace("2013.csv","2013")
pop_df = pop_df.replace("2014.csv","2014")
pop_df = pop_df.replace("2015.csv","2015")
pop_df = pop_df.replace("2016.csv","2016")
pop_df = pop_df.replace("2017.csv","2017")
pop_df = pop_df.replace("2018.csv","2018")

pop_df = pop_df.sort_values(["Name","Year"])
pop_df = pop_df.reset_index(drop=True)
pop_df.head(10)
```

```python
pop_df = pop_df.drop(columns=['state'])
pop_df
```

```python
states_code = pd.read_csv("states_code.csv")
states_code.head()
```

```python
final_pop = pop_df.merge(states_code, left_on='Name', right_on='State')
final_pop.head()
```

```python
final_pop = final_pop.drop(columns=['Abbrev','State'])
```

```python
final_pop = final_pop.rename(columns={"Code":"State"})
final_pop
```

```python
final_pop.to_csv("total_population.csv",index = None, header=True)
```
